```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
invLoss0 = []
with open("inv loss0", "r") as f:
    for line in f:
        loss = float(line)
        invLoss0.append(loss)
```

```python
invLoss = []
with open("inv loss", "r") as f:
    for line in f:
        loss = float(line)
        invLoss.append(loss)
```
